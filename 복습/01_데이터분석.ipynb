```python
#데이터 프레임 df의 특정 열이 가지고 있는 고유값 확인
unique_values=df['origin'].value_counts()
unique_values
```

```python
#cylinders수를 기준으로 데이터들의 평균 확인
df.groupby(["cylinders"],as_index=False)['weight'].mean()
```

```python
#cylinders수를 기준으로 중량, 가속시간 평균
df.groupby(["cylinders"], as_index=False)[['weight','acceleration']].mean()
```

```python
#년도, cylinders수를 기준으로 데이터의 평균 확인
df.groupby(["modelyear","cylinders"],as_index=False)['weight'].mean()
```

```python
#년도, cylinders수를 기준으로 데이터의 평균,중앙값 확인
df.groupby(["modelyear","cylinders"],as_index=False)['weight'].agg(['mean','median'])
```

```python
#데이터에 들어간 ?를 0으로 변경
df['horsepower']=df['horsepower'].replace('?',0)
```

```python
#데이터타입 변경
df['horsepower']=df['horsepower'].astype('float')
df.info()
```

```python
#데이터타입 확인
df['horsepower']=df['horsepower'].value_counts()
df.info()
```
