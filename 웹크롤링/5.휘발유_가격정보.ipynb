## 실습2) HTML TABLE데이터 가져오기

### 휘발유 가격정보 가져오기

- 네이버 → 증권 → 하단의 유가 더보기 → 국내 휘발유 → 일별시세 → 페이지에서 새창

#### 1 페이지만 가져오기

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs4
import warnings
warnings.filterwarnings('ignore')

url = 'https://finance.naver.com/marketindex/oilDailyQuote.naver'
p = {'marketindexCd' : 'OIL_GSL', 'page' : '1'}
res = requests.get(url, params=p)
soup = bs4(res.text, 'html.parser')
soup
```

```python
table = soup.select('.tbl_exchange.today')
```

```python
df = pd.read_html(str(table))[0] # 데이터 프레임만 바깥으로 꺼낼려고 0을 사용
df
```

#### 20 페이지 가져오기

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs4
import warnings
warnings.filterwarnings('ignore')

url = 'https://finance.naver.com/marketindex/oilDailyQuote.naver'
result = []
for i in range(1,20):
    param = {'marketindexCd' : 'OIL_GSL', 'page' : i}
    res = requests.get(url, params=param)
    soup = bs4(res.text, 'html.parser')
    table = soup.select('.tbl_exchange.today')
    df = pd.read_html(str(table))[0] # 데이터 프레임만 바깥으로 꺼낼려고 0을 사용
    result.append(df)
    
df_result=pd.concat(result, ignore_index=True)
display(df_result)
```
