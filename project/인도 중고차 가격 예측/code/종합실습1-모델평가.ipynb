```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
```

```python
df_train_1, df_test = train_test_split(df_train, test_size = 0.3, random_state = 1234) 
print("train data size : {}".format(df_train_1.shape))
print("test data size : {}".format(df_test.shape))
```

```python
df_x = df_train.drop(columns="Price")
df_y = df_train['Price']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_numeric = df_x.select_dtypes(include="number")  # 숫자형 변수만 선택
df_x[df_numeric.columns] = scaler.fit_transform(df_x[df_numeric.columns])

df_raw_dummy = pd.get_dummies(df_x)
df_raw_y = df_y
df_raw_x = df_raw_dummy

df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y
                                                                , test_size = 0.3, random_state = 1234) 
```

```python
# 다중회귀분석
reg_model = smf.ols(formula = "Price ~ C(Location)+Kilometers_Driven+\
                    +Seats+Engine_CC+Power_bhp+C(Brand)+\
                    C(Year_cat)+C(car_cat)+Mil_cat_mean+\
                    +C(new_Year)", data = df_train_1)
reg_result = reg_model.fit()



# 의사결정트리
tree_graph = DecisionTreeRegressor(random_state=1234, min_samples_leaf=8,
                                   min_samples_split=15, max_depth = 9,max_features='sqrt' )
tree_graph.fit(df_train_x, df_train_y)

# 랜덤포레스트
rf_model = RandomForestRegressor(max_depth=15, min_samples_leaf=5, min_samples_split=5, 
                                  n_estimators=150, random_state=1234)
rf_model.fit(df_train_x, df_train_y)

# 그래디언트 부스팅
gb_final = GradientBoostingRegressor(random_state=1234, n_estimators = 100, min_samples_leaf = 1,
                                     min_samples_split = 4, max_depth = 4, learning_rate = 0.2)
gb_final.fit(df_train_x, df_train_y)

# XGBOOST
xgb_model = XGBRegressor(learning_rate=0.1,max_depth= 5,n_estimators= 200,reg_alpha= 0.1,reg_lambda= 0.5, random_state=1234)
xgb_model.fit(df_train_x, df_train_y)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

models = ["다중회귀분석","의사결정트리", "랜덤 포레스트", "그래디언트 부스팅", 'XGBOOST']
mse_list, rmse_list, mae_list, mape_list = [], [], [], []
r2_score_list = []

# 다중회귀분석 평가
test_pred = reg_result.predict(df_test)
mse_list.append(mean_squared_error(df_test['Price'], test_pred))
rmse_list.append(np.sqrt(mse_list[-1]))
mae_list.append(mean_absolute_error(df_test['Price'], test_pred))
mape_list.append(np.mean(np.abs((df_test['Price'] - test_pred) / df_test['Price'])) * 100)
r2_score_list.append(r2_score(df_test['Price'], test_pred))

# 의사결정트리 평가
dt_y_pred = tree_graph.predict(df_test_x)
mse_list.append(mean_squared_error(df_test_y, dt_y_pred))
rmse_list.append(np.sqrt(mse_list[-1]))
mae_list.append(mean_absolute_error(df_test_y, dt_y_pred))
mape_list.append(np.mean(np.abs((df_test_y - dt_y_pred) / df_test_y)) * 100)
r2_score_list.append(r2_score(df_test_y, dt_y_pred))
    
# 랜덤 포레스트 평가
rf_y_pred = rf_model.predict(df_test_x)
mse_list.append(mean_squared_error(df_test_y, rf_y_pred))
rmse_list.append(np.sqrt(mse_list[-1]))
mae_list.append(mean_absolute_error(df_test_y, rf_y_pred))
mape_list.append(np.mean(np.abs((df_test_y - rf_y_pred) / df_test_y)) * 100)
r2_score_list.append(r2_score(df_test_y, rf_y_pred))

# 그래디언트 부스팅 평가
gb_y_pred = gb_final.predict(df_test_x)
mse_list.append(mean_squared_error(df_test_y, gb_y_pred))
rmse_list.append(np.sqrt(mse_list[-1]))
mae_list.append(mean_absolute_error(df_test_y, gb_y_pred))
mape_list.append(np.mean(np.abs((df_test_y - gb_y_pred) / df_test_y)) * 100)
r2_score_list.append(r2_score(df_test_y, gb_y_pred))

# XGBoost 평가
xgb_y_pred = xgb_model.predict(df_test_x)
mse_list.append(mean_squared_error(df_test_y, xgb_y_pred))
rmse_list.append(np.sqrt(mse_list[-1]))
mae_list.append(mean_absolute_error(df_test_y, xgb_y_pred))
mape_list.append(np.mean(np.abs((df_test_y - xgb_y_pred) / df_test_y)) * 100)
r2_score_list.append(r2_score(df_test_y, xgb_y_pred))
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728','#9467bd']

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].bar(models, mse_list, color=colors)
ax[0, 0].set_title("mse").set_fontsize(15)
for i, v in enumerate(mse_list):
    ax[0, 0].text(i, v + 0.01, str(round(v, 3)), ha='center', va='bottom')
ax[0, 0].tick_params(axis='x', rotation=45)

ax[0, 1].bar(models, rmse_list, color=colors)
ax[0, 1].set_title("rmse").set_fontsize(15)
for i, v in enumerate(rmse_list):
    ax[0, 1].text(i, v + 0.01, str(round(v, 3)), ha='center', va='bottom')
ax[0, 1].tick_params(axis='x', rotation=45)

ax[1, 0].bar(models,mae_list, color=colors)
ax[1, 0].set_title("mae").set_fontsize(15)
for i, v in enumerate(mae_list):
    ax[1, 0].text(i, v + 0.01, str(round(v, 3)), ha='center', va='bottom')
ax[1, 0].tick_params(axis='x', rotation=45)

ax[1, 1].bar(models, mape_list, color=colors)
ax[1, 1].set_title("mape").set_fontsize(15)
for i, v in enumerate(mape_list):
    ax[1, 1].text(i, v + 0.01, str(round(v, 3)), ha='center', va='bottom')
ax[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
