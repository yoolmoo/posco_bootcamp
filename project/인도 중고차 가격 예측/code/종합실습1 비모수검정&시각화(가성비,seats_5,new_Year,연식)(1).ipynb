#### 비모수 검정 방법

    - Mann-Whitney U 검정 (Wilcoxon 순위 합 검정): 두 개의 독립적인 그룹 간의 중앙값 차이를 비교합니다.
    - Wilcoxon 부호 순위 검정: 두 개의 대응하는 그룹 간의 중앙값 차이를 비교합니다.
    - Kruskal-Wallis H 검정: 세 개 이상의 독립적인 그룹 간의 중앙값이 동일한지 비교합니다.
    - Friedman 검정: 세 개 이상의 대응하는 그룹 간의 중앙값이 동일한지 비교합니다.
    - Kolmogorov-Smirnov 검정: 두 표본의 분포를 비교합니다.
    - Runs Test: 데이터의 무작위성을 검정합니다.
    - Sign Test: 대응하는 표본에서 중앙값이 0인지 비교합니다.

## 1. 가성비

- Wilcoxon 부호 순위 : 두 개의 대응하는 그룹 간의 중앙값 차이를 비교합니다.

```python
from scipy.stats import wilcoxon

# '가성비' 열의 값이 'good'인 데이터와 'bad'인 데이터 분리
good_data = df_train[df_train['가성비'] == 'good']['Price']
bad_data = df_train[df_train['가성비'] == 'bad']['Price']

# 두 집단의 샘플 크기 확인
print("Good data size:", len(good_data))
print("Bad data size:", len(bad_data))

# 두 집단의 샘플 크기를 동일하게 맞추기 위해 샘플 크기 조정
min_size = min(len(good_data), len(bad_data))
good_data = good_data.sample(min_size, random_state=42)
bad_data = bad_data.sample(min_size, random_state=42)

# Wilcoxon 부호 순위 검정 수행
statistic, p_value = wilcoxon(good_data, bad_data)

# 결과 출력
print("Wilcoxon statistic:", statistic)
print("p-value:", p_value)
```

# 2. Seats_5

 - Kruskal-Wallis H 검정: 세 개 이상의 독립적인 그룹 간의 중앙값이 동일한지 비교합니다.

```python
from scipy.stats import kruskal

# 각 그룹의 데이터 준비
group1 = df_train[df_train['Seats_5'] == '5 좌석']['Price']
group2 = df_train[df_train['Seats_5'] == '5 초과 좌석']['Price']
group3 = df_train[df_train['Seats_5'] == '5 미만 좌석']['Price']

# Kruskal-Wallis H 검정 수행
statistic, p_value = kruskal(group1, group2, group3)

# 결과 출력
print("<Kruskal-Wallis H 검정 결과>")
print("검정 통계량 (H-statistic):", statistic)
print("p-value:", p_value)
```
