```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel("1월.xlsx")
df2 = pd.read_excel("2월.xlsx")
df3 = pd.read_excel("3월.xlsx")
df4 = pd.read_excel("4월.xlsx")
```

```python
df = pd.concat([df1,df2,df3,df4], axis=0)
df.drop(columns="구간", inplace=True)
df.rename(columns={'요일' : '일자'}, inplace=True)
```

```python
df.replace('-', np.nan , inplace=True)
for i in range(24):
    df[i].fillna(df[i].mean().round(1), inplace=True)
```

```python
df_sum = df.groupby('일자').sum()
df_sum = df_sum.astype(int)
df_sum['요일'] = df_sum.index.dayofweek
df_sum
```

```python
df_ex = df_sum.copy()
```

```python
df_ex.reset_index(inplace=True)
df_ex.rename(columns={'index': '일자'}, inplace=True)

# '일자' 열을 기준으로 피벗하여 데이터 재구성
df_pivoted = pd.melt(df_ex, id_vars=['일자','요일'], var_name='시간', value_name='교통량')

# '일자-시간' 열 생성
df_pivoted['일시'] = df_pivoted['일자'].astype(str) + '-' + df_pivoted['시간'].astype(str)

# 필요없는 열 제거
df_result = df_pivoted.drop(columns=['일자', '시간'])
```

```python
df_result
```

```python
df_result['일시'] = df_result['일시'] + '시'
df_result.set_index('일시', inplace=True)
df_result.index = pd.to_datetime(df_result.index, format='%Y-%m-%d-%H시')
df_result = df_result.sort_index()
```

```python
df_result
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.family'] = 'NanumGothic'

matplotlib.rc("axes", unicode_minus = False)
```

```python
df_result['일'] = df_result.index.day
df_result['시'] = df_result.index.strftime('%H')
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

train_start_date = '2024-01-23'
train_end_date = '2024-04-11'
test_start_date = '2024-04-12'
test_end_date = '2024-04-12'

train_data = df_result[train_start_date:train_end_date]
test_data = df_result[test_start_date:test_end_date]

X_test = test_data[['시','일','요일']]
y_test = test_data['교통량'].to_frame()

label = '교통량'  
predictor = TabularPredictor(label=label, problem_type='regression', eval_metric='rmse').fit(train_data)

# train 데이터를 사용하여 예측
y_pred = predictor.predict(X_test)

from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
X_test['실제교통량'] = y_test
X_test['예측교통량'] = y_pred
X_test
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X_test.index, X_test['실제교통량'], label='실제 교통량')
plt.plot(X_test.index, X_test['예측교통량'], label='예측 교통량')
plt.xlabel('일시')
plt.ylabel('교통량')
plt.title('시간에 따른 실제 교통량 및 예측 교통량')
plt.xticks(rotation=45)  # 20개의 눈금만 표시하도록 설정
plt.legend()
plt.show()
```

```python
X_test.drop(columns="실제교통량",inplace=True)
X_test.rename(columns={'예측교통량': '교통량'}, inplace=True)
```

```python
train_start_date = '2024-04-10 00:00:00'
train_end_date = '2024-04-11 23:00:00'

train_data1 = df_result[train_start_date:train_end_date]
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# train_data1, X_test를 합친 combined 데이터 생성
combined = pd.concat([train_data1, X_test])

plt.figure(figsize=(10, 6))

# 전체 데이터를 파란색 선으로 플롯
plt.plot(combined.index, combined['교통량'], label='전체 교통량', color='blue')

# X_test에 해당하는 부분을 빨간색으로 강조하여 플롯
plt.plot(X_test.index, X_test['교통량'], label='예측 교통량', color='red')

plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=6))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H시'))

plt.title('교통량 예측 시각화', fontsize=16)
plt.xlabel('일시', fontsize=14)
plt.ylabel('교통량', fontsize=14)
plt.legend(loc='upper left', fontsize=14)  # label 위치 조정 및 글자 크기 조정
plt.xticks(fontsize=11, rotation=45)
plt.show()
```
