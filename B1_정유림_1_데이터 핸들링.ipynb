# 실습5

### step1 : 데이터 구성하기

```python
df4 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
df5 = pd.DataFrame({'E':['E4','E5','E6','E7'],
                  'F':['F4','F5','F6','F7'],
                  'G':['G4','G5','G6','G7'],
                  'H':['H4','H5','H6','H7']})
df6 = pd.DataFrame({'I':['I8','I9','I10','I11'],
                  'J':['J8','J9','J10','J11'],
                  'K':['K8','K9','K10','K11'],
                  'L':['L8','L9','L10','L11']})
```

```python
df4
```

```python
df5
```

```python
df6
```

### step2: concat을 이용한 데이터 결합

```python
#concat을 이용해 df4,5,6 가로로 결합 => axis=1 : 가로로 결합해라
result = pd.concat([df4,df5,df6],axis=1)
#concat 결과 확인
result
```

### step3: parameter변경

```python
#ignore_index=True: 기존 index가 아닌 새로운 index로 재생성
result=pd.concat([df4,df5,df6],axis=1,ignore_index=True)
#concat결과 데이터의 변수명 생성
result.columns=["col_"+str(i) for i in range(12)]
#concat 결과 데이터 result확인 => 변수명이 col_1,col_2,,이런식으로 잘 변경됨
result
```
